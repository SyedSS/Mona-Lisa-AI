# How many humans are in space right now?

### Querying the International Space Station (http://open-notify.org/)


```python
# Import packages

import requests
import json
from datetime import date
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
# Find location of ISS

response = requests.get("http://api.open-notify.org/iss-now.json")
print(response.text)
```

```python
# Extract the latitude and longitude

iss_location = response.json()['iss_position']
lat = iss_location['latitude']
lon = iss_location['longitude']
print('Latitude:', lat)
print('Longitude:', lon)
```

```python
# Plot the current location of the ISS

plt.figure(figsize=(8, 8))
m = Basemap(projection='ortho', resolution=None, lat_0=lat, lon_0=lon)
m.bluemarble(scale=0.5);
x, y = m(lon, lat)
plt.plot(x, y, marker="p", color='yellow',markersize=15)
plt.text(x, y, '  The ISS Location',color='yellow',fontsize=15);
```

```python
# Find out how many humans in space

internationalSpaceStation = "http://api.open-notify.org/astros.json"
inSpace = requests.get(internationalSpaceStation).json()['number']
print("Today's date:", date.today())
print("The Number of Humans in Space Right Now:", inSpace)
```
